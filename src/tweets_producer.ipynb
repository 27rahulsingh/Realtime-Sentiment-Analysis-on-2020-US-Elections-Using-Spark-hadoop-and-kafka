```python
import pandas as pd
from time import sleep
from kafka import KafkaProducer
import json
```

```python
def json_serializer(data):
    return json.dumps(data).encode("utf-8")
```

```python
producer = KafkaProducer(bootstrap_servers = ['chinmay:9092'],
                         value_serializer=json_serializer)
```

```python
df1 = pd.read_csv("data/hashtag_donaldtrump.csv",lineterminator ='\n')
df2 = pd.read_csv("data/hashtag_joebiden.csv", lineterminator='\n')
```

```python
filtered_df = df1[(df1['country'] == 'United States of America') & (~df1['state'].isnull())]

# Create a new DataFrame with only the desired columns
df1 = filtered_df[['tweet', 'country', 'state']]

# Display the first few rows of the new DataFrame
df1.head(10)
```

```python
filtered_df = df2[(df2['country'] == 'United States of America') & (~df2['state'].isnull())]

# Create a new DataFrame with only the desired columns
df2 = filtered_df[['tweet', 'country', 'state']]

# Display the first few rows of the new DataFrame
df2.head()
```

```python
df1.shape
```

```python
df2.shape
```

```python
for index, (row1, row2) in enumerate(zip(df1.iterrows(), df2.iterrows())):
    # Extract tweet, country, and state information from df1
    tweet1 = row1[1]['tweet']
    state1 = row1[1]['state']

    tweet2 = row2[1]['tweet']
    state2 = row2[1]['state']
    
    # Send tweet, country, and state information from df1 to Kafka
    sleep(0.1)
    producer.send('trump', {'tweet': tweet1, 'state': state1})
    
    # Send tweet, country, and state information from df2 to Kafka
    producer.send('biden', {'tweet': tweet2, 'state': state2})  # Assuming country and state information is not available in df2
    
    if index > 100000:
        break
```

```python
producer.flush()
```
